##Agregation

```python
df = pd.DataFrame({
    'x': [3.1, 5.5, 9.2, 1.7, 1.2, 8.3, 2.6],
    'y': [1.4, np.nan, 5.0, 5.8, 9.0, np.nan, 9.2]
})
df.agg('sum')
df.agg(np.sum)
def foofunc(x):
    return x.iloc[1]
df.agg(foofunc)
df.agg(['sum', 'mean'])
df.agg({'x': ['sum', 'mean'], 'y': ['min', 'max']})
df.agg('min', axis=1)
def nth_value(x, n):
    return x.iloc[n]
df.agg(nth_value, n=3)
df.agg(func=lambda x, n: x.iloc[n], n=3)
```

##Groupby

```python
from colorsys import yiq_to_rgb
generator = np.random.default_rng(123)
df = pd.DataFrame({
    'john': generator.choice([True, False], size=10, replace=True),
    'judy': generator.choice([True, False], size=10, replace=True)
})

def days_til_party(x):
    z = np.array(x)
    y = np.array([])
    flag = 1
    for i in z:
      if i:
          flag = 0
          c = 0
      if flag: 
        y = np.append(y, np.nan)
      else:
        y = np.append(y, c)
        c += 1       
    return pd.Series(y, dtype=pd.Int64Dtype())
       

df['party'] = df.john & df.judy
df['reversed_party'] = df.party.sort_index(ascending = False, ignore_index=True)
df['days_til_party'] = df.reversed_party.transform(lambda x: days_til_party(x)).sort_index(ascending = False, ignore_index=True)

print(df)
```

```python
potholes = pd.DataFrame({
    'length':[5.1, np.nan, 6.2, 4.3, 6.0, 5.1, 6.5, 4.3, np.nan, np.nan],
    'width':[2.8, 5.8, 6.5, 6.1, 5.8, np.nan, 6.3, 6.1, 5.4, 5.0],
    'depth':[2.6, np.nan, 4.2, 0.8, 2.6, np.nan, 3.9, 4.8, 4.0, np.nan],
    'location':pd.Series(['center', 'north edge', np.nan, 'center', 'north edge', 'center', 'west edge',
                          'west edge', np.nan, np.nan], dtype='string')
})
# Given your DataFrame of pothole measurements, discard rows where more than half the values are NaN, 
# elsewhere impute NaNs with the average value per column unless the column is non-numeric,
# in which case use the mode.

potholes = potholes.apply({'length' : (lambda x: potholes.length.mean() if pd.isna(x) else x),
                            'width' : (lambda x: potholes.width.mean() if pd.isna(x) else x),
                            'depth' : (lambda x: potholes.depth.mean() if pd.isna(x) else x),
                            'location' : (lambda x: potholes.location.mode()[0] if pd.isna(x) else x)})

print(potholes)
```
