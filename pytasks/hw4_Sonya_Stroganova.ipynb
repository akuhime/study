## Problem 1

Напишите функцию `is_prime`, проверяющую целое число n на простоту.

```python
from math import sqrt
def is_prime(n):
    for i in range(2, round(sqrt(n))+1):
      if n % i == 0:
        return False
    return True
```

```python
lst_ = []
for i in range(2,100):
  if is_prime(i):
    lst_.append(i)

print(*lst_)
```

## Problem 4

Напишите генератор, возвращающий бесконечную последовательность чисел из треугольника Паскаля. Последовательность, возвращаемая генератором, должна выглядеть так:
`1 1 1 1 2 1 1 3 3 1 1 4 6 4 1 1 5 10 10 5 1 1 6 15 20 15 6 1 1 7 21 35 35 21 7 1 1 8 28 56 70 56 28 8 1 1 9 36 84 126 126 84 36 9 1 ...`

```python
from math import factorial

def C(n, k):
  return int(factorial(n)/(factorial(k)*factorial((n-k))))

def gen_pascal():
  c = 0
  while True:
    for i in range(c+1):
      res = C(c, i)
      yield res
    c += 1

it = gen_pascal()
lst_ = []
for i in range(55):
  lst_.append(next(it))
print(*lst_)
```

## Problem 5

Написать генератор `limit(generator, max_count)`.
Возвращает не более `max_count` значений генератора `generator`

```python
def limit(generator, max_count):
    count = 0 
    it = iter(generator)
    while count < max_count:
      yield next(it)
      count += 1
```

```python
# for test
it = limit(range(10), 5)
for value in it:  # для тестов взят range, хоть он и не является генератором
    print(value, end=' ')
  
#print(next(it))
```

## Problem 6

Написать генератор `all_elements(list)`.
Возвращает все элементы списка `list` любой вложенности.

Указание: для проверки того, что объект итерируемый, можно проверить, что он наследник `Iterable` (используйте функции `isinstance`/`issubclass`  и модуль `collections`)

```python
from collections import Iterable
def all_elements(lst_):
  for i in lst_:
    if isinstance(i, Iterable):
      for j in all_elements(i):
        yield j 
    else:
      yield i
```

```python
# for test
values = [1, [2, 3], [4, [5, 6], [[[7]]]], 8]
list(all_elements(values)) == [1, 2, 3, 4, 5, 6, 7, 8]
```

## Problem 7* (20 %)

Напишите генератор `merge_sorter`, выполняющий слияние отсортированных последовательностей целых чисел.

Генератор принимает произвольное число аргументов. Аргументом может быть любой итерируемый объект, в том числе другой генератор. Гарантируется, что каждый аргумент — это последовательность целых чисел, отсортированных в порядке неубывания.

```python
def merge_sorter(*args):
    dict_ = {}
    flag = 0
    c = 0
    args = list(map(iter, args))
    for i in args:
      c += 1
      dict_[c] = next(i)
    while not flag:
      min_ = min(list(dict_.values()))
      yield min_
      for key in dict_.keys():
        if dict_[key] == min_:
          index = key
      try:
        dict_[index] = next(args[index-1])
      except StopIteration:
        dict_.pop(index)
      # print(dict_.values())
      if dict_ == {}:
        flag = 1
```

```python
it_1 = filter(lambda x: x % 3 == 0, (i for i in range(10)))
it_2 = filter(lambda x: x % 3 == 1, (i for i in range(10)))
it_3 = filter(lambda x: x % 3 == 2, (i for i in range(10)))

it = merge_sorter(list(it_1),it_2,it_3)
for i in it:
  print(i, end = ' ')
```

Пользоваться глобальными переменными запрещено.
Декоратор должен вести себя порядочно, то есть не должен затирать документацию функции.

Проверить работу ваших декораторов стоит на функции [Аккермана](https://ru.wikipedia.org/wiki/%D0%A4%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D1%8F_%D0%90%D0%BA%D0%BA%D0%B5%D1%80%D0%BC%D0%B0%D0%BD%D0%B0) или на функции, принимающей в себя произвольное число списков и склеивающей их в один (правда для этого реализовать их придется самостоятельно :) ). Аккермана удобно проверять с аргументами (3,4)).

В этой задаче обнуляю число вызовов с помощью sys._getframe(1).f_code.co_name

Это мне кажется плохим решением, так как я не знаю что такое frame и не могу предсказать вывод sys._getframe(n)

Каким образом можно этого избежать??

```python
from datetime import datetime
import sys 

def profiler(func):
  def wrapper(*args, **kwargs):
    if not hasattr(wrapper, 'calls'):
      setattr(wrapper, 'calls', 0) 
    else:
      wrapper.calls += 1
    if sys._getframe(1).f_code.co_name == '<module>': 
      wrapper.calls = 0
    start = datetime.now()
    res =  func(*args, **kwargs)
    wrapper.last_time_taken = datetime.now() - start
    return res
  wrapper.__name__ = func.__name__
  wrapper.__doc__ = func.__doc__
  wrapper.__module__ = func.__module__
  return wrapper

@profiler
def accerman(m, n):
  if m == 0:
    return n + 1
  elif n == 0:
    return accerman(m-1, 1)
  else:
    return accerman(m-1, accerman(m, n-1))

def print_accerman(m, n):
  print(f'm = {m}, n = {n}')
  print(f'value of accerman function is {accerman(m, n)}')
  print(f'accerman.calls is {accerman.calls}')
  print(f'accerman.last_time_taken is {accerman.last_time_taken}')
  print()

print_accerman(3, 4)
print_accerman(1, 4)
```
