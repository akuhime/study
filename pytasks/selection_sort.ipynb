сортировка выбором

```python
def mysort(lst):
  n = len(lst)
  for i in range(n-1):
    k = i
    m = lst[i]
    for j in range(i+1, n):
      if lst[j] < m:
        k = j
        m = lst[j]
    lst[i], lst[k] = lst[k], lst[i]
  return lst

print(mysort([8, 6, 9, 2, 1, 5, 0, 4, 9, 0, 0, 10, 13, 2]))
```

Мат ожидание числа проверок 
$$\frac{1}{p}\cdot(1-(1-p)^{n})$$
Где $p$ вероятность что символ равен искомому, $n$ длина последовательности



```python
def func(n, p=0.33):
  return ((1-pow(1-p, n))/p)
```

```python
import matplotlib.pyplot as plt
x = [i for i in range(2, 100)]
y = list(map(func, x))
plt.scatter(x, y)
```

```python
import matplotlib.pyplot as plt
x = [i for i in range(2, 100000)]
y = [func(i, 1/i) for i in range(2, 100000)]
plt.scatter(x, y)
```
