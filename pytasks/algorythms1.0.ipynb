# Дано a,b,c вывести в порядке возрастания все корни ax^2+bx+c = 0

```python
import math
a, b, c = list(map(int, input().split()))
d = b*b - 4*a*c
if a == 0:
  if not b == 0:
    print(-c/b)
  elif b == 0 and c == 0:
    print('infinite number of solutions') 
else:
  if d == 0:
    print((-b)/(2*a))
  elif d > 0:
    tuple_ = ((-b-math.sqrt(d))/(2*a), (-b+math.sqrt(d))/(2*a))
    print(*sorted(tuple_))
```

# Удаление (замена) символов в строке

```python
s = ')-+i((m+ a+ s)+-t++r-+i()n)g'
trans_table = str.maketrans({'+' : '', '-' : '', '(' : '', ')' : ''})
print(s.translate(trans_table))
```

# Любимая рекурсия

```python
N, K, M = list(map(int, input().split()))

def func(N, K, M, S=0):
  D = (N // K)*(K // M)
  if D == 0:
    return S
  else:
    S += D
    N2 = N - D*M
    return func(N2, K, M, S)

print(func(N, K, M))
```

# Юзаю  itertools

```python
import itertools
a, b, c, d, e = list(map(int, input().split()))
flag = 0
for i, j in list(itertools.permutations([a, b, c], 2)):
  if i <= d and j <= e:
    print('YES')
    flag = 1
    break
if not flag:
  print('NO')
```

```python
import re
s = 'AAABBZXTTTREEEEBBBXXXXYYJJJKKKKKKKKKKKKKLLLLLLAS'
def func(s):
  if s == '':
      pass
  elif re.fullmatch(r'[A-Z]*', s):
    prev_elem = s[0]
    counter = 0
    print(s[0], end = '')
    for now_elem in s:
      if now_elem == prev_elem:
        counter += 1
      else:
        if counter > 1:
          print(counter, end = '')
        counter = 1
        print(now_elem, end = '')
      prev_elem = now_elem
    if counter > 1:
          print(counter, end = '')
  else:
    print('bebebe')
func(s)
```

```python
import itertools
N, M, K = [int(i) for i in input().split()]
lst_ = [[0 for i in range(M)] for j in range(N)]

for i in range(K):
  a, b = [int(i) for i in input().split()]
  lst_[a-1][b-1] = '*'

meow = list(itertools.product([-1,0,1], [-1,0,1]))

for i in range(0, N):
  for j in range(0, M):
    counter = 0
    if lst_[i][j] == '*':
      pass
    else:
      for s in meow:
        if (i+s[0] < 0) or (i+s[0] > N-1) or (j+s[1] < 0) or (j+s[1] > M-1):
          pass
        else:
          if lst_[i+s[0]][j+s[1]] == '*':
            counter += 1
      lst_[i][j] = counter

for s in lst_:
  for t in s:
    print(t, end = ' ')
  print()
```
